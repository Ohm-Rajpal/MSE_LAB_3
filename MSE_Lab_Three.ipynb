```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

# Data Collected in Lab

```python
hardness_data = [36, 33, 43, 46, 49, 48, 48, 48, 48, 48, 41, 30, 22, 11, 4, 4, 5, 0, -15]
data = {
    'Position': list(range(1, 20)),
    'Major Load (kg)': [60] * 19,
    'Minor Load (kg)': [10] * 19,
    'Hardness': hardness_data
}

data = pd.DataFrame(data)
data
```

# Graph For Question 1

```python
# Collecting the hardness values
hardness_values = list(data['Hardness'][:5])
percent_values = [0.10, 0.20, 0.30, 0.40, 0.50]
# Plot
plt.plot(percent_values, hardness_values)
plt.xlabel('Percent Reduction in Thickness')
plt.ylabel('Rockwell A Hardness Measurement')
plt.title('Rockwell Hardness vs Percentage Reduction')
# Display
plt.show()
```

# Question 3 Hardness and Length

```python
hardness_values = list(data['Hardness'][4:])
corresponding_length = np.arange(0, 1.875, 1/8)
# Plot points
plt.scatter(corresponding_length, hardness_values)
plt.xlabel('Corresponding Length in inches')
plt.ylabel('Rockwell Hardness Values in HRA')
plt.title('Hardness vs Length')
# Show
plt.show()
```
